# CONTROL SET: keeping 2 patients per category out as control 

```python
# exclude from the training 3 patients per label ( they have 3 run at least per patience)
def extract_unique_subjects(df):
    extracted_subjects = {}
    for label, group in df.groupby('disease'):
        unique_subjects = group[~group['subject'].isin(extracted_subjects.keys())]['subject'].unique()[:2]
        subjects_runs = group[group['subject'].isin(unique_subjects)][['subject', 'Run']]
        extracted_subjects[label] = subjects_runs.groupby('subject')['Run'].apply(list).to_dict()
    return extracted_subjects

subjects_to_exclude = extract_unique_subjects(meta_data)
#ie = [val for sublist in subjects_to_exclude['HIV Broad Neutralizing'].values() for val in sublist]
subjects_to_exclude
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combine all data into a DataFrame for easier plotting
data = pd.DataFrame({
    'v_identity':  bnabs_v_identity['v_identity'].tolist() + nobnabs_v_identity['v_identity'].tolist() +   no_v_identity['v_identity'].tolist(),
    'Group':  ['BRO'] * len(bnabs_v_identity) +   ['NEU'] * len(nobnabs_v_identity) + ['NON'] * len(no_v_identity)
})


# === Custom colors ===
custom_palette = {
    'nAb': '#A6CEE3',  # light blue
    'bnAb': '#005a86ff',   # dark blue
    'nnAb': '#43B02A'
}

# === Plot style ===
plt.rcParams.update({
    "font.size": 16,
    "axes.titlesize": 18,
    "axes.labelsize": 16,
    "xtick.labelsize": 15,
    "ytick.labelsize": 15,
    "legend.fontsize": 14,
})

# === Create violin plot ===
plt.figure(figsize=(8.5, 4))
sns.set_style("white")

label_map = {'NEU': 'nAb', 'BRO': 'bnAb', 'NON': 'nnAb'}
data['Group_label'] = data['Group'].map(label_map)

order = ['nnAb', 'nAb', 'bnAb']

violin = sns.violinplot(
    x='Group_label',
    y='v_identity',
    data=data,
    palette=custom_palette,  # assicurati che chiavi corrispondano
    order=order
)


# Set alpha manually
for patch in violin.collections:
    patch.set_alpha(1.0)

# === Axis and labels ===
plt.ylabel('V gene identity (%)', fontsize=16)
plt.xlabel('')
plt.grid(False)
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

# === Add average value and count per group ===
avg_values = data.groupby('Group_label')['v_identity'].mean()
for i, group in enumerate(['nnAb', 'nAb', 'bnAb']):
    avg = avg_values[group]
    count = len(data[data['Group_label'] == group])
    # Adjust horizontal positioning: small offset to avoid overlap

    plt.text(i + 0.23, 63, f"Avg: {avg:.2f}", ha='center', fontsize=13,
             bbox=dict(facecolor='white', edgecolor='none', alpha=0.8))
    plt.text(i + 0.23, 54, f"#: {count}", ha='center', fontsize=13,
             bbox=dict(facecolor='white', edgecolor='none', alpha=0.8))

# === Save figure ===
plt.tight_layout()
plt.savefig(out_path + 'PAPER_v_identity_cdrh3_unique_withavg.png', dpi=300, bbox_inches='tight')
plt.savefig(out_path + 'PAPER_v_identity_cdrh3_unique_withavg.svg', dpi=300, bbox_inches='tight')
plt.show()
```
